- clean and prepare Data

```python
# Drop features that doesn't valuable in my data
df.drop(['LinkedIn URL (optional)','Date of birth','Facebook URL','Any questions?'],axis=1,inplace=True)
```

```python
df.rename(columns={'Full Name (3 names at least)':'full_Name' , 'Faculty - Department':'Fac_Dep',
                   'What do you know about Z Fast Racing Team?':'Zfast_knowing','How did you hear about us?':'know_Us?',
                   'Do you have any volunteering experience?':'have_vol_exp?','If yes, what is your volunteering experience?':'vol_exp',
                   'What do you know about electrical vehicle ?':'elec_veh_knowing' ,'What do you know about competition we participate ?':'comp_part_knowing',
                   'What do you seek to learn by joining Z Fast Team?':'seek_to_learn'} , inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.set_index("full_Name" ,inplace=True)  #  make names is my index as non duplicated values 
```

```python
df["elec_veh_knowing"].fillna("no experiance" ,inplace=True)
df["comp_part_knowing"].fillna("don't know",inplace=True)
```
