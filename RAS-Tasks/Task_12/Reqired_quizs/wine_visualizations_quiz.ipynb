# Plotting with Matplotlib
Use Matplotlib to create bar charts that visualize the conclusions you made with groupby and query.

```python
# Import necessary packages and load `winequality_edited.csv`
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv("winequality_edited.csv")
```

### #1: Do wines with higher alcoholic content receive better ratings?
Create a bar chart with one bar for low alcohol and one bar for high alcohol wine samples.

```python
# Use query to select each group and get its mean quality
median = df['alcohol'].median()
low = df.query(f'alcohol < {median}')
high = df.query(f'alcohol >= {median}')

quality_low_M = low['quality'].mean()
quality_high_M = high['quality'].mean()
```

```python
# Create a bar chart with proper labels
po_s = [1, 2]
heights = [quality_low_M, quality_high_M]
labels = ['Low', 'High']

plt.bar(po_s, heights, tick_label=labels)

plt.xlabel('Alcohol ')
plt.ylabel('Average Quality ');
```

### #2: What level of acidity receives the highest average rating?
Create a bar chart with a bar for each of the four acidity levels.

```python
# Use groupby to get the mean quality for each acidity level
mean_of_acidity = df.groupby('acidity_levels').quality.mean()
```

```python
# Create a bar chart with proper labels
po_s = [1,2,3,4]
height = mean_of_acidity.values
labels = mean_of_acidity.index
plt.bar( po_s, height  ,   tick_label=labels)

plt.xlabel('Acidity Levels')
plt.ylabel('Quality ');
```
